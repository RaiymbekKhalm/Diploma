```python
import os
import matplotlib.pyplot as plt
```

```python
root_folder = "C:/Users/User/Desktop/diploma/Dataset/wholeset/train"
image_counts = []
```

```python
for folder_name in os.listdir(root_folder):
    folder_path = os.path.join(root_folder, folder_name)
    if os.path.isdir(folder_path) and not folder_name.startswith('.'):
        count = len([f for f in os.listdir(folder_path) if f.endswith('.jpg') or f.endswith('.png') or f.endswith('.jpeg')])
        image_counts.append(count)
```

```python
plt.bar(range(len(image_counts)), image_counts, tick_label=os.listdir(root_folder))
plt.title('Гистограмма объема данных в датасете')
plt.show()
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
def create_image_data_generators(data_dir, input_shape, batch_size):
    datagen = ImageDataGenerator(rescale=1.0/255, validation_split=0.2)

    train_generator = datagen.flow_from_directory(
        data_dir,
        target_size=input_shape[:2],
        batch_size=batch_size,
        class_mode='categorical',
        subset='training',
        shuffle=True
    )

    validation_generator = datagen.flow_from_directory(
        data_dir,
        target_size=input_shape[:2],
        batch_size=batch_size,
        class_mode='categorical',
        subset='validation',
        shuffle=True
    )

    return train_generator, validation_generator
```
